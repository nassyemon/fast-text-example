```python
from gensim.models import KeyedVectors
import os
```

```python
# model_filepath = './data/entity_vector.model.bin'
model_filepath = './data/entity_vector.model.bin'
model = KeyedVectors.load_word2vec_format(model_filepath, binary=True)
```

```python
l = model.most_similar(["アイドル", "オタク"], topn=30)

list(filter(lambda x: not x[0].startswith("["), l))
```
